# Sample Notebook

This notebook demonstrates basic data processing for conversion testing.

```python
import json

data = [
    {"name": "Alice", "score": 95},
    {"name": "Bob", "score": 87},
    {"name": "Carol", "score": 72},
]

print(f"Records: {len(data)}")
print(json.dumps(data, indent=2))
```

## Analysis

Compute the average score and find the top performer.

```python
avg = sum(d["score"] for d in data) / len(data)
top = max(data, key=lambda d: d["score"])

print(f"Average score: {avg:.1f}")
print(f"Top performer: {top['name']} ({top['score']})")
```
